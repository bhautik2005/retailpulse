#**# FEATURE ENGINEERING**

Order-Level Features

```python
df = merged_df.copy()

# Total order value
df['total_value'] = df['price'] + df['freight_value']

# Delivery time (days)
df['delivery_time'] = (
    df['order_delivered_customer_date'] - df['order_purchase_timestamp']
).dt.days

# Approval time (processing delay)
df['approval_time'] = (
    df['order_approved_at'] - df['order_purchase_timestamp']
).dt.seconds / 3600  # hours
```

Delivery Performance Features

```python
# Delivery delay (late or early)
df['delivery_delay'] = (
    df['order_delivered_customer_date'] - df['order_estimated_delivery_date']
).dt.days

# Late delivery flag
df['is_late'] = (df['delivery_delay'] > 0).astype(int)
```

Time-Based Features

```python
df['order_year'] = df['order_purchase_timestamp'].dt.year
df['order_month'] = df['order_purchase_timestamp'].dt.month
df['order_day'] = df['order_purchase_timestamp'].dt.day
df['order_weekday'] = df['order_purchase_timestamp'].dt.weekday
df['order_hour'] = df['order_purchase_timestamp'].dt.hour
```

Customer-Level Aggregation (RFM BASE 🔥)

```python
customer_df = df.groupby('customer_unique_id').agg({
    'order_purchase_timestamp': ['max', 'count'],
    'payment_value': 'sum',
    'review_score': 'mean'
}).reset_index()

customer_df.columns = [
    'customer_id',
    'last_purchase',
    'frequency',
    'monetary',
    'avg_review'
]
```

Calculate Recency

```python
import datetime

latest_date = df['order_purchase_timestamp'].max()

customer_df['recency'] = (
    latest_date - customer_df['last_purchase']
).dt.days
```

RFM Segmentation

```python
# Recency score
r_cut = pd.qcut(customer_df['recency'], 4, duplicates='drop')
r_labels = list(range(len(r_cut.cat.categories), 0, -1)) if len(r_cut.cat.categories) > 0 else []
customer_df['R_score'] = pd.qcut(customer_df['recency'], 4, labels=r_labels, duplicates='drop')

# Frequency score
f_cut = pd.qcut(customer_df['frequency'], 4, duplicates='drop')
f_labels = list(range(1, len(f_cut.cat.categories) + 1)) if len(f_cut.cat.categories) > 0 else []
customer_df['F_score'] = pd.qcut(customer_df['frequency'], 4, labels=f_labels, duplicates='drop')

# Monetary score
m_cut = pd.qcut(customer_df['monetary'], 4, duplicates='drop')
m_labels = list(range(1, len(m_cut.cat.categories) + 1)) if len(m_cut.cat.categories) > 0 else []
customer_df['M_score'] = pd.qcut(customer_df['monetary'], 4, labels=m_labels, duplicates='drop')

customer_df['RFM_score'] = (
    customer_df['R_score'].astype(str) +
    customer_df['F_score'].astype(str) +
    customer_df['M_score'].astype(str)
)
```

Product-Level Features (Inventory + Demand)

```python
product_df = df.groupby('product_id').agg({
    'order_id': 'count',
    'payment_value': 'sum',
    'price': 'mean'
}).reset_index()

product_df.columns = [
    'product_id',
    'total_orders',
    'total_revenue',
    'avg_price'
]
```

Daily Sales (FOR FORECASTING)

```python
time_series_df = df.groupby(
    df['order_purchase_timestamp'].dt.date
).agg({
    'payment_value': 'sum'
}).reset_index()

time_series_df.columns = ['date', 'daily_sales']
```
